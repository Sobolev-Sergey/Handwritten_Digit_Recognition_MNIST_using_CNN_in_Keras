# Классификация рукописных цифр

MNIST - набор данных, для оценки моделей машинного обучения в задаче классификации рукописных цифр. Каждое изображение представляет собой квадрат 28 на 28 пикселей (всего 784 пикселя). Стандартное разделение набора данных используется для оценки и сравнения моделей, где 60 000 изображений используются для обучения модели, а отдельный набор из 10 000 изображений используется для ее тестирования. Есть 10 цифр (от 0 до 9) или 10 классов для прогнозирования

## Загрузка набора данных MNIST в Keras

```python
# Plot ad hoc mnist instances
from keras.datasets import mnist
import matplotlib.pyplot as plt

# load (downloaded if needed) the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

```

## Примеры из train набора данных MNIST

```python
# plot 12 images as gray scale
for i in range(1,13):
    plt.subplot(2,6,i)
    plt.imshow(X_train[i-1], cmap=plt.get_cmap('gray'))

# show the plot
plt.show()
```

## Примеры из test набора данных MNIST

```python
# plot 12 images as gray scale
for i in range(1,13):
    plt.subplot(2,6,i)
    plt.imshow(X_test[i-1], cmap=plt.get_cmap('gray'))

# show the plot
plt.show()
```
